```python
import calendar
from datetime import datetime

def get_week_number(x):
    """Parses date in 'MM/DD/YYYY' format and returns ISO week number"""
    # %m = Month, %d = Day, %Y = 4-digit Year
    date_obj = datetime.strptime(x, "%m/%d/%Y") 
    return date_obj.isocalendar()[1]
```

```python
df_train_final['week_number']=df_train_final['date'].apply(lambda x:get_week_number(x))
df_test_final['week_number']=df_test_final['date'].apply(lambda x:get_week_number(x))
df_val_final['week_number']=df_val_final['date'].apply(lambda x:get_week_number(x))
```

```python
def get_season(x):
    """This function is used to get season in US according to various months"""
    if x in [12,1,2]:
        return 0      #"Winter"
    elif x in [3,4,5]:
        return 1   #"Spring"
    elif x in [6,7,8]:
        return 2   #"Summer"
    else:
        return 3   #"Autumn"
```

```python
df_train_final['season']=df_train_final['month'].apply(lambda x:get_season(x))
df_test_final['season']=df_test_final['month'].apply(lambda x:get_season(x))
df_val_final['season']=df_val_final['month'].apply(lambda x:get_season(x))
```

```python
def check_if_quater_begin(x):
    """This is used to check if day is beginning of quarter"""
    # Parse the date string "MM/DD/YYYY"
    date_obj = datetime.strptime(x, "%m/%d/%Y")
    
    day = date_obj.day
    month = date_obj.month
    
    # Returns 1 if it is the 1st of Jan, Apr, Jul, or Sep
    return 1 if (day == 1 and (month in [1, 4, 7, 10])) else 0
```

```python
df_train_final['quater_start']=df_train_final['date'].apply(lambda x:check_if_quater_begin(x))
df_test_final['quater_start']=df_test_final['date'].apply(lambda x:check_if_quater_begin(x))
df_val_final['quater_start']=df_val_final['date'].apply(lambda x:check_if_quater_begin(x))
```

```python
def check_if_quater_end(x):
    """This is used to check if day is beginning of quarter"""
    # Parse the date string "MM/DD/YYYY"
    date_obj = datetime.strptime(x, "%m/%d/%Y")
    
    day = date_obj.day
    month = date_obj.month
    
    # Returns 1 if it is the 1st of Jan, Apr, Jul, or Sep
    if (day==31 and month==3) or (day==30 and month==6) or (day==30 and month==9) or (day==31 and month==12):
        return 1
    else:
        return 0
```

```python
df_train_final['quater_end']=df_train_final['date'].apply(lambda x:check_if_quater_end(x))
df_test_final['quater_end']=df_test_final['date'].apply(lambda x:check_if_quater_end(x))
df_val_final['quater_end']=df_val_final['date'].apply(lambda x:check_if_quater_end(x))
```

```python
def month_start(x):
    """This is used to check if day is begining of month"""
    date_obj = datetime.strptime(x, "%m/%d/%Y")
    day = date_obj.day
    return 1 if day==1 else 0
```

```python
df_train_final['month_start']=df_train_final['date'].apply(lambda x:month_start(x))
df_test_final['month_start']=df_test_final['date'].apply(lambda x:month_start(x))
df_val_final['month_start']=df_val_final['date'].apply(lambda x:month_start(x))
```

```python
def month_end(x):
    """This is used to check if day is end of month"""
    date_obj = datetime.strptime(x, "%m/%d/%Y")
    
    day = date_obj.day
    month = date_obj.month
    year = date_obj.year
    leap_yr=(year%4==0) # to check if it is a leap year
    val=(day==31 and month==1) or (day==29 if leap_yr else day==28) or (day==31 and month==3) or (day==30 and month==4) or\
        (day==31 and month==5) or (day==30 and month==6) or (day==31 and month==7) or (day==31 and month==8) or\
        (day==30 and month==9) or (day==31 and month==10) or (day==30 and month==11) or (day==31 and month==12)
    return 1 if val else 0
```

```python
df_train_final['month_end']=df_train_final['date'].apply(lambda x:month_end(x))
df_test_final['month_end']=df_test_final['date'].apply(lambda x:month_end(x))
df_val_final['month_end']=df_val_final['date'].apply(lambda x:month_end(x))
```

```python
def year_start(x):
    """This is used to check if day is begining of year"""
    date_obj = datetime.strptime(x, "%m/%d/%Y")
    
    day = date_obj.day
    month = date_obj.month
    return 1 if (day==1 and month==1) else 0

def year_end(x):
    """This is used to check if day is end of year"""
    date_obj = datetime.strptime(x, "%m/%d/%Y")
    
    day = date_obj.day
    month = date_obj.month
    return 1 if (day==31 and month==12) else 0
```

```python
df_train_final['year_start']=df_train_final['date'].apply(lambda x:year_start(x))
df_test_final['year_start']=df_test_final['date'].apply(lambda x:year_start(x))
df_val_final['year_start']=df_val_final['date'].apply(lambda x:year_start(x))

df_train_final['year_end']=df_train_final['date'].apply(lambda x:year_end(x))
df_test_final['year_end']=df_test_final['date'].apply(lambda x:year_end(x))
df_val_final['year_end']=df_val_final['date'].apply(lambda x:year_end(x))
```

```python
import gc
import numpy as np
```

```python
tt = pd.concat([df_train_final, df_val_final, df_test_final], ignore_index=True)

tt.sort_values(['item_id', 'date'], inplace=True)

df_wide = tt.pivot_table(index=['item_id'], 
                         columns='date', 
                         values='sales')

df_wide.fillna(0, inplace=True)

del tt
gc.collect()

df_wide
```

```python
train = df_train_final.copy()
validation = df_val_final.copy()
test = df_test_final.copy()
```

```python
agg_funcs = ['mean', 'std']
shifts = [28]
windows = [7, 14, 30, 60, 360]


for aggregate in agg_funcs:
    for shif in shifts:
        for r in windows:
            roll = df_wide.rolling(r, axis=1).agg(aggregate).shift(shif)
            
            dates = roll.columns
            feature_name = f"roll_{r}_shift_{shif}_{aggregate}"
            
            roll = roll.astype('float16')
            
            roll.reset_index(inplace=True)
            
            roll = pd.melt(roll, 
                           id_vars=['item_id'], 
                           value_vars=dates, 
                           var_name='date', 
                           value_name=feature_name)
            
            roll.fillna(-1, inplace=True)
            
            
            train = train.merge(roll, on=['item_id', 'date'], how='left')
            validation = validation.merge(roll, on=['item_id', 'date'], how='left')
            test = test.merge(roll, on=['item_id', 'date'], how='left')
            
            del roll
            gc.collect()

train
```

```python
alphas = [0.99] #0.95, 0.5
SHIFT_DAY = 28

for alpha in alphas:
    roll = df_wide.shift(SHIFT_DAY, axis=1).ewm(alpha=alpha, axis=1, adjust=False).mean()
    
    dates = roll.columns
    feature_name = f"ewm_alpha_{str(alpha).replace('.','')}_shift_{SHIFT_DAY}"
    
    roll = roll.astype('float16')
    
    roll.reset_index(inplace=True)
    
    roll = pd.melt(roll, 
                   id_vars=['item_id'], 
                   value_vars=dates, 
                   var_name='date', 
                   value_name=feature_name)
    
    roll.fillna(-1, inplace=True)
    
    train = train.merge(roll, on=['item_id', 'date'], how='left')
    validation = validation.merge(roll, on=['item_id', 'date'], how='left')
    test = test.merge(roll, on=['item_id', 'date'], how='left')
    
    del roll
    gc.collect()
```

```python
for lag in range(28, 100, 7):
    feat_name = 'direct_lag_' + str(lag)
    
    # 1. Shift the wide dataframe (columns are dates)
    # We shift by 'lag' columns to the right
    lag_i = df_wide.shift(lag, axis=1)
    
    # 2. Capture the date columns before resetting index
    # (These are the columns we will melt)
    dates = lag_i.columns
    
    # 3. Handle the Index to prepare for melt
    # Move 'item_id' from index to a column
    lag_i.reset_index(inplace=True)
    
    # 4. Melt from Wide to Long format
    # This aligns the data so it can be merged with your training sets
    lag_i = pd.melt(lag_i, 
                    id_vars=['item_id'], 
                    value_vars=dates, 
                    var_name='date', 
                    value_name=feat_name)
    
    # 5. Fill NaNs and optimize memory
    lag_i.fillna(-1, inplace=True)
    lag_i[feat_name] = lag_i[feat_name].astype('int16')
    
    # 6. Merge the new feature into your 3 datasets
    # We use 'how=left' to ensure we don't lose rows if dates don't perfectly align
    train = train.merge(lag_i, on=['item_id', 'date'], how='left')
    validation = validation.merge(lag_i, on=['item_id', 'date'], how='left')
    test = test.merge(lag_i, on=['item_id', 'date'], how='left')
    
    print("Feature created for lag", lag)
    
    # 7. Clean up to prevent RAM explosion
    del lag_i
    gc.collect()

train
```
